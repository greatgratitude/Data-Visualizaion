```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# to create a figure with a single axes

x = [1, 2, 3]
y = [4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show() #to display the plot
```

## creating a figure with multiple subplots

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[0].set_title("Left Plot")
axes[1].plot(x, y)
axes[1].set_title("Right Plot")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, y)
axes[0, 1].plot(x, y)
axes[1, 0].plot(x, y)
axes[1, 1].plot(x, y)
plt.show()
```

```python
df = pd.read_csv(r"C:\Users\T O S H I B A\Downloads\netflix_titles.csv")

df.head()
```

```python
df.info()
```

## what are the numbers of netflits titles released per year: is there any trend?

```python
titles_per_year = df.groupby("release_year")["title"].count().reset_index()

titles_per_year
```

```python
plt.plot(titles_per_year["release_year"], titles_per_year["title"], color="red", marker='o')
plt.title("Number of Netflix Titles Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Nunber of Titles")


plt.show()
```

```python
type_counts = df["type"].value_counts()
```

```python
type_counts.index
```

```python
type_counts
```

```python
plt.pie(type_counts, labels=type_counts.index, autopct='%1.2f%%', colors=["navy", "lightblue"])
plt.title("Distribution of Movies and TV Shows")
plt.show()
```

### to clean the data on rating column

```python
df["rating"] = df["rating"].replace({"74 min":"Unknown", "84 min":"Unknown", "66 min":"Unknown"})
```

```python
df["rating"].value_counts()
```

```python
rating_count = df["rating"].value_counts()
```

```python
rating_count.values
```

```python
rating_count.index
```

```python
plt.bar(rating_count.index, rating_count.values)
plt.xticks(rotation=90)
plt.title("Number of titles per Rating")
plt.xlabel("Rating")
plt.ylabel("Number of titles")
plt.show()
```

## Histogram

```python
plt.hist(df["release_year"], bins=50)
plt.title("Distribution of release years")
plt.show()
```

```python
#movies and tv shows released per year

release_year_by_type = df.groupby(['release_year', 'type']).size()

release_year_by_type
```

```python
release_year_by_type = df.groupby(['release_year', 'type']).size().unstack(fill_value=0).reset_index()

release_year_by_type
```

```python
plt.plot(release_year_by_type["release_year"], release_year_by_type["Movie"], label = "Movie")


plt.plot(release_year_by_type["release_year"], release_year_by_type["TV Show"], label = "TV Show")

plt.legend()
plt.title("Movies and TV Shows released over the years")

plt.show()
```
